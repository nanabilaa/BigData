```python
import pandas as pd
from google.colab import files

uploaded = files.upload()

# Load dataset
dataset = pd.read_csv('iris.csv', header=None)  # dataset tidak ada header
dataset.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species'] #buat header dataset
print(dataset.info())

# Step 1: Missing Values
missing_values = dataset.isnull().sum()
print("\nMissing values per column:")
print(missing_values)

# Step 2: Normalisasi Kolom Numerik
numerical_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Perhitungan Min-Max Scaling
for column in numerical_columns:
    dataset[column] = (dataset[column] - dataset[column].min()) / (dataset[column].max() - dataset[column].min())

# Step 3: Setelah Normalisasi
print("\nDataset Setelah Normalisasi:")
print(dataset.head())
dataset['species'] = dataset['species'].str.lower().str.strip()

# Hapus duplikasi
dataset = dataset.drop_duplicates()
print(dataset.head())
```

```python
#13
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files
uploaded = files.upload()

file_name = list(uploaded.keys())[0]
iris_data = pd.read_csv(file_name)

iris_data.columns = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Species'] # beri header

# 1. Scatter plot SepalLength AND SepalWidth
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris_data, x='SepalLength', y='SepalWidth', hue='Species', palette='viridis')
plt.title('Sepal Length AND Sepal Width')
plt.show()

#1.2 PetalLength AND PetalWidth
plt.figure(figsize = (8,6))
sns.scatterplot(data=iris_data, x='PetalLength', y='PetalWidth')
plt.title('PETALLENGTH AND PETAL WIDTH')
plt.show()

# 2. Histogram Distribusi sepal length
plt.figure(figsize=(8, 6))
sns.histplot(iris_data['SepalLength'], kde=True, color='blue')
plt.title('Distribution Sepal Length')
plt.xlabel('Sepal Length')
plt.show()

# 3. Bar plot: Distribusi Spesies
plt.figure(figsize=(8, 6))
sns.countplot(data=iris_data, x='Species', palette='pastel')
plt.title('Distribusi spesies bunga iris')
plt.xlabel('Spesies')
plt.ylabel('Perhitungan')
plt.show()
```
